# Phase 4: Visualization & Result Analysis

Generate charts, word clouds, and extract insights for the report.

```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path('../data')
REPORT_DIR = DATA_DIR / 'reports'

train = pd.read_parquet(DATA_DIR / 'train_clean.parquet')
val = pd.read_parquet(DATA_DIR / 'val_clean.parquet')
test = pd.read_parquet(DATA_DIR / 'test_clean.parquet')
all_data = pd.concat([train, val, test], ignore_index=True)
print(f'Total: {len(all_data)} rows')
```

```python
import sys
sys.path.insert(0, str(Path('../scripts').resolve()))
from visualize import plot_sentiment_pie, plot_sentiment_bar, plot_topic_bar, plot_word_clouds, plot_sentiment_by_source, extract_insights

plot_sentiment_pie(all_data, 'Sentiment Distribution (Overall)', 'nb_sentiment_pie.png')
plot_sentiment_bar(all_data, 'Sentiment Count', 'nb_sentiment_bar.png')
uit = all_data[all_data['source'] == 'uit_vsfc']
plot_topic_bar(uit, 'nb_topic_bar.png')
plot_word_clouds(all_data, 'nb_word_clouds')
plot_sentiment_by_source(all_data, 'nb_sentiment_by_source.png')
print('Charts generated.')
```

```python
insights = extract_insights(all_data)
print('Sentiment %:', insights['sentiment_percentages'])
print('Top topic by sentiment:', insights.get('top_topic_by_sentiment', {}))
print('Top negative words:', list(insights.get('top_negative_words', {}).keys())[:15])
print('Top positive words:', list(insights.get('top_positive_words', {}).keys())[:15])
```
