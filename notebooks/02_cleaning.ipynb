# Phase 2: Data Cleaning & Augmentation

This notebook demonstrates the noise injection and Vietnamese text cleaning pipeline.

```python
from pathlib import Path
import sys

import pandas as pd

sys.path.insert(0, str(Path('../app').resolve()))
from cleaning import clean_text

DATA_DIR = Path('../data')
```

## 1. Load data and show noisy samples vs clean samples

```python
uit = pd.read_parquet(DATA_DIR / 'uit_vsfc_noisy.parquet')
noisy = uit[uit['is_noisy']].head(10).copy()
noisy['clean'] = noisy['text'].apply(lambda t: clean_text(t, do_segment=False))
noisy[['sentiment', 'text', 'clean']]
```

## 2. Comparison table: before vs after cleaning

```python
noisy['before_len'] = noisy['text'].str.len()
noisy['after_len'] = noisy['clean'].str.len()
noisy['tokens'] = noisy['clean'].str.split().str.len()

print('=== Stats on 10 noisy samples ===')
print(f'Avg chars before: {noisy["before_len"].mean():.0f}')
print(f'Avg chars after:  {noisy["after_len"].mean():.0f}')
print(f'Avg tokens after: {noisy["tokens"].mean():.1f}')
```

## 3. Clean dataset statistics

```python
train = pd.read_parquet(DATA_DIR / 'train_clean.parquet')
val = pd.read_parquet(DATA_DIR / 'val_clean.parquet')
test = pd.read_parquet(DATA_DIR / 'test_clean.parquet')

print(f'Train: {len(train)}')
print(f'Val:   {len(val)}')
print(f'Test:  {len(test)}')
print()
for name, split in [('Train', train), ('Val', val), ('Test', test)]:
    print(f'{name} sentiment: {split["sentiment"].value_counts().to_dict()}')
```

```python
all_data = pd.concat([train, val, test], ignore_index=True)
all_data['source'].value_counts()
```
